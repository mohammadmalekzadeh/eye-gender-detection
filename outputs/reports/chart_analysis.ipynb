### Importing librayr

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read data and set style

```python
sns.set(style="whitegrid")

df = pd.read_csv('../reports/model_evaluation_results.csv')

metrics = df.columns.to_list()[1:6]
```

### Show plot

```python
for metric in metrics:
    plt.Figure(figsize=(10, 6))
    sns.barplot(data=df, x='Model', y=metric, palette="Set2")
    plt.title(f"{metric} Comparison Across Models", fontsize=16)
    plt.ylim(0, 1)
    plt.xlabel("Models")
    plt.ylabel(metric)
    plt.xticks(rotation=30)
    plt.tight_layout()
    plt.savefig(f"../reports/plots/{(metric.lower())}_comparison.png")
    plt.show()
```

### Combined plot

```python
df_melted = df.drop(columns='Confusion Matrix').melt(id_vars="Model", var_name="Metric", value_name="Score")

plt.Figure(figsize=(12,7))

sns.lineplot(data=df_melted, x="Model", y="Score", hue="Metric", marker="o", palette="Set2", linewidth=2.5)

plt.title("Model Evalution Metrics Comparison", fontsize=16)
plt.xlabel("Models", fontsize=12)
plt.ylabel("Score", fontsize=12)
plt.xticks(rotation=30)
plt.ylim(0, 1.05)
plt.legend(title="Metric")
plt.tight_layout()
plt.savefig("../reports/plots/combined_metrics_comparison")
plt.show()
```
